```python
import pandas as pd

pop = pd.read_csv('state-population.csv')
areas=pd.read_csv('state-areas.csv')
abbrevs=pd.read_csv('state-abbrevs.csv')

print(pop.head())
print(areas.head())
print(abbrevs.head())

merged=pd.merge(pop,abbrevs, how='outer',left_on='state/region',right_on='abbreviation')
print(merged.head())

merged=merged.drop(['abbreviation'],axis=1)
print(merged)

print(merged.isnull().any())

print(merged[merged['population'].isnull()].head())

print(merged.loc[merged['state'].isnull(),'state/region'].unique())

merged.loc[merged['state/region']=='PR','state']='Puerto Rico'
merged.loc[merged['state/region']=='USA','state']='United States'
print(merged.isnull().any())

final=pd.merge(merged,areas,on='state',how='left')
print(final.head())

final.isnull().any()

final['density']=final['population']/final['area (sq. mi)']
print(final)

final.dropna(inplace=True)
final.head()

final.isnull().any()


## 2011년 주별로 18세이하 인구밀도가 높은 상위 5개주
## California, Alabama, Arizona, Arkansasm, Alaska
data2011=final.query("year==2011 & ages== 'under18'")
print(data2011.head())
print(data2011.head().sort_values(by='density', ascending=False))

data2010=final.query("year==2010 & ages== 'under18'")
print(data2010.head())

data2011=final.query("year==2011 & ages=='under18'")
print(data2011.head())

data2012=final.query("year==2012 & ages=='under18'")
print(data2012.head())
```

```python
df=pd.merge(data2010,data2011, on='state', how='left')
df.head()
```

```python
df=pd.merge(df,data2012, on='state',how='left')
df.head()
```

```python
df
```

```python
df["increase2011"]=(df["density_y"]-df["density_x"])/df["density_x"]*100
df
```

```python
df["increase2012"]=(df["density"]-df["density_y"])/df["density"]*100
df
```

```python
## 18세 이하 인구밀도가 2010년 < 2011년 < 2012년으로 증가세가 이루어진 주
# Colorado,District of Columbia,Florida,Hawaii,North Dakotam,Nebraska, Oklahoma,South Dakota,Texas,Utah,Virginia,Washington,Wyoming
result=df.query("increase2011>0 & increase2012>0")
print(result)
print(result.state)
```

```python
density2011 = data2011.density.values
print(density2011)
```

```python
density2012=data2012.density.values
print(density2012)
```

```python
rate2011=(density2011-density2010)/density2010*100
rate2011=pd.DataFrame({'rate2011':rate2011})
print(rate2011)
```

```python
import numpy as np
rate2012=(density2012-density2011)/density2011*100
rate2012=pd.DataFrame({'rate2012':rate2012})
print(rate2012)
```

```python
increase=pd.concat([rate2011,rate2012], axis=1)
increase
```

```python
final=pd.concat([final,increase],axis=1)
print(final)
```
